```python
filtered_lotsize=df[df['Lotsize']<30000] #narrowing down to properties with lot sizes of < 30,000 ft
filtered_lot_price=filtered_lotsize[filtered_lotsize['Sale_price']<0.5e6] 
#showing properties with sale prices less than $500,000

p1=sns.regplot(x='Lotsize',y='Sale_price', data=filtered_lot_price, order=1, line_kws={'color':'red'})

slope, intercept, r, p1, sterr = scipy.stats.linregress(x=p1.get_lines()[0].get_xdata(),
                                                        y=p1.get_lines()[0].get_ydata())
#Finding the slope, intercept, correlation (r) and standard error of the linear regression of the data.

plt.text(31000,100000,'y = ' + str(round(intercept,3)) + ' + ' + str(round(slope,3)) + 'x' +', r = ' + str(round(r,3)) ) 
plt.show()
```

```python
filtered_lot_price_copy = filtered_lot_price.copy()
filtered_lot_price_copy['Style'] = filtered_lot_price_copy['Style'].astype(str)

print(filtered_lot_price_copy['Style'].dtype)
```

```python
unique_styles = filtered_lot_price_copy['Style'].unique()
print(sorted(unique_styles))
#after printing the sorted values of Style, we find a nan value at the end
```

```python
filtered_lot_price_copy['Style']=filtered_lot_price_copy['Style'].fillna('None').astype(str)
#Replace all nan values with blank values, then convert them to the string value type
print(filtered_lot_price_copy['Style'].dtype)
print(sorted(filtered_lot_price_copy['Style'].unique()))
```

```python
filtered_lot_price_copy_unique=sorted(filtered_lot_price_copy['Style'].unique())
print(filtered_lot_price_copy_unique)
```

```python
# Replace NaN values with 'None' using numpy's where function
filtered_lot_price_copy['Style'] = np.where(pd.isna(filtered_lot_price_copy['Style']), 'None', filtered_lot_price_copy['Style']).astype(str)

#pd.isna(filtered_lot_price_copy['Style'] DOES NOT SOLVE THIS PROBLEM, APPARENTLY...
# Print the data type
print(filtered_lot_price_copy['Style'].dtype)

# Print sorted unique styles
sorted_unique_styles = sorted(filtered_lot_price_copy['Style'].unique())
print(sorted_unique_styles)
```

```python
# Replace 'NaN' values (case-insensitive) with 'None'. THIS being case-insensitive solved the problem with 'nan' values
filtered_lot_price_copy['Style'] = filtered_lot_price_copy['Style'].str.replace('NaN', 'None', case=False)

# Convert the column to string type
filtered_lot_price_copy['Style'] = filtered_lot_price_copy['Style'].astype(str)

# Print the data type
print(filtered_lot_price_copy['Style'].dtype)

# Print sorted unique styles
sorted_unique_styles = sorted(filtered_lot_price_copy['Style'].unique())
print(sorted_unique_styles)
```
