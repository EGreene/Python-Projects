```python
#I am dropping fields about doubles tournaments winners
df1=df.drop(['doubles_winner_1_name','doubles_winner_1_url','doubles_winner_1_player_slug','doubles_winner_1_player_id',
            'doubles_winner_2_name','doubles_winner_2_url','doubles_winner_2_player_slug','doubles_winner_2_player_id'],axis=1)
#I display all columns in the headers
pd.set_option('display.max_columns', None)
df1.head()
```

```python
pd.set_option('display.max_columns', None)
#Shows the number of tournaments won by player name
df1.value_counts("singles_winner_name")
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
singles_and_tournaments=df1.value_counts(subset=['singles_winner_name','tourney_name'])
#Shows the players and tournaments won by tournament name. Players who won the most of a specific tournament are shown first.
singles_and_tournaments
```

```python
majors_won_by_tourn=singles_and_tournaments[singles_and_tournaments.index.get_level_values('tourney_name').isin(['Wimbledon','US Open',
                                                                                                                 'Australian Open', 'Roland Garros'])]
print(majors_won_by_tourn)
```

```python
mask=df1['tourney_name'].str.contains('Australian Open|Roland|Wimbledon|US Open', case=False, regex=True)
matching_rows=df1[mask]
matching_rows1=matching_rows.value_counts('tourney_name')
#Shows the number of tournaments that have the names of the grand slams in the tournament name. This helps me find "Australian Open-2"
matching_rows1
```

```python
#A way to get an index with the grand slam tournament names
matching_rows1.index.get_level_values('tourney_name').unique()
```

```python
#I had to include "Australian Open-2" because it turns out that the Australian Open was sometimes played in alternate dates and this is recorded
#with a different tournament name in the csv file
#Before, Vitas Gerulaitis was listed as not having won major, but he won the Australian Open in Dec 1977 when it took place on a different date 
#than usual.
tournaments_to_exclude = ['Wimbledon', 'US Open', 'Roland Garros', 'Australian Open','Australian Open-2']
```

```python
#Shows the names of all players who won tournaments
players_who_won_any = singles_and_tournaments.index.get_level_values('singles_winner_name').unique()
print(players_who_won_any)
```

```python
players_who_never_won_excluded = [player for player in players_who_won_any if not singles_and_tournaments[player, :].
                                  index.get_level_values('tourney_name').isin(tournaments_to_exclude).any()]
#Here I used AI to figure out how to get this data. This would be much easier to find in SQL, for instance.
#This shows all players who won tournaments but no grand slams.
players_who_never_won_excluded
```

```python
filtered_result = singles_and_tournaments.loc[players_who_never_won_excluded, :]
#This shows a list of all players who won tournaments but never won a grand slam, sorted by most tournaments won.
filtered_result.groupby(['singles_winner_name']).count().reset_index(name='Count').sort_values(['Count'], ascending=False)
```
